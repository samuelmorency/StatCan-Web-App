```python
imm_df = pd.read_excel("IIC_CSBP_D3_Enrollments-Graduates_2019-2022.xlsx", sheet_name="IMM")
imm_df.head()
```

```python
def clean_data(df):
    # Drop columns: 'CIP6 Code', 'Province or Territory' and 4 other columns
    df_clean = (df
          .drop(columns=['CIP6 Code', 'Province or Territory', 'Enrollment Type', 'CMAname', 'CSDname'])
          .rename(columns={'Source Sheet': 'STEM/BHASE'})
          .astype({' PR code': 'category', 
                   'ISCED Level of Education': 'category', 
                   'Credential Type': 'category', 
                   'CIP Name': 'category', 
                   'PRdguid': 'category', 
                   'Academic Year': 'category', 
                   'STEM/BHASE': 'category', 
                   'Institution': 'category', 
                   'CMAuid': 'category', 
                   'CSDuid': 'category'})
    )
    df_clean['STEM/BHASE'] = (df_clean['STEM/BHASE']
                        .str.replace("^BHASE\\w+?$", "BHASE", case=False, regex=True)
                        .str.replace("^STEM\\w+?$", "STEM", case=False, regex=True)
    )
    df_clean['Value'] = df_clean['Value'].replace("..", np.nan).astype('float').astype('Int32')

    return df_clean

gen_df_clean = clean_data(gen_df)
imm_df_clean = clean_data(imm_df)
gen_df_clean.head()
```

```python
province_sf = gpd.read_file("hidden/shapes/lpr_000b21a_e.shp")
cma_sf = gpd.read_file("hidden/shapes/lcma000b21a_e.shp")
csd_sf = gpd.read_file("hidden/shapes/lcsd000b21a_e.shp")

to_convert = [province_sf, cma_sf, csd_sf]
converted = [df.to_crs(epsg=4326) for df in to_convert]
province_sf, cma_sf, csd_sf = converted

# Optimize data cleaning with chained operations
cma_longlat_clean = (cma_sf
    .rename(columns={'CMANAME': 'NAME'})
    .drop(columns=['DGUIDP', 'CMAUID', 'CMAPUID','CMATYPE','LANDAREA'])
    #.assign(CLASS="CMA_CA")
)

csd_longlat_clean = (csd_sf
    .rename(columns={'CSDNAME': 'NAME'})
    .drop(columns=['CSDUID','CSDTYPE','LANDAREA'])
    #.assign(CLASS="CSD")
)

combined_longlat = pd.concat([cma_longlat_clean, csd_longlat_clean], ignore_index=True)
combined_longlat_indexed = combined_longlat#.set_index('DGUID')
combined_longlat_clean = combined_longlat_indexed[combined_longlat_indexed['DGUID'].isin(dguids)]
combined_longlat_clean.to_parquet("combined_longlat_clean.parquet")
combined_longlat_clean
```
