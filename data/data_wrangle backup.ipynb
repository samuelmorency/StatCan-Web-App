```python
gen_df = pd.concat([gen_df_19_22, gen_df_22_23], ignore_index=True)
gen_df['Academic Year'] = gen_df['Academic Year'].astype(str).str.replace('_', '/')
gen_df
```

```python
def clean_data(df):
    # Drop columns: 'CIP6 Code', 'Province or Territory' and 4 other columns
    df_clean = (df
          .drop(columns=['CIP6 Code', ' PR code', 'Enrollment Type', 'CMAname', 'CSDname'])
          .rename(columns={'Source Sheet': 'STEM/BHASE',
                           'PRdguid': 'PRDGUID',
                           'CMAuid': 'CMAPUID',
                           'CSDuid': 'CSDUID'})
          #.astype({'Province or Territory': 'category', 
                   #'ISCED Level of Education': 'category', 
                   #'Credential Type': 'category', 
                   #'CIP Name': 'category', 
                   #'PRDGUID': 'category', 
                   #'Academic Year': 'category', 
                   #'STEM/BHASE': 'category', 
                   #'Institution': 'category', 
                   #'CMAPUID': 'category', 
                   #'CSDUID': 'category', 
                   #'DGUID': 'category'})
    )
    df_clean['GRAD/ENROLL'] = df_clean['STEM/BHASE'].str.replace("^[^_]+_[^_]+_([^_]+)_.*$", lambda match: match.group(1), regex=True)
    df_clean['GEO'] = df_clean['STEM/BHASE'].str.replace(".*_([^_]*)$", lambda match: match.group(1), regex=True)
    df_clean['STEM/BHASE'] = (df_clean['STEM/BHASE']
                        .str.replace("^BHASE\\w+?$", "BHASE", case=False, regex=True)
                        .str.replace("^STEM\\w+?$", "STEM", case=False, regex=True)
    )
    df_clean = df_clean[df_clean['Value'] != ".."]
    df_clean['Value'] = df_clean['Value'].astype('Int32')

    return df_clean

gen_df_clean = clean_data(gen_df)
gen_df_clean
```

```python

# Filter out rows with 'GRAD/ENROLL' as 'ENROLL'
gen_df_grads = gen_df_clean[gen_df_clean['GRAD/ENROLL'] == 'GRAD'].drop(columns=['GRAD/ENROLL'])
gen_df_grads = gen_df_grads[gen_df_grads['CMAPUID'] != 'Z_Non-CMACA']
gen_df_grads = gen_df_grads[gen_df_grads['GEO'] != 'PROV']
gen_df_grads = gen_df_grads.drop(columns=['PRDGUID', 'CMAPUID', 'CSDUID', 'GEO'])

# Remove rows with missing Value
gen_df_grads = gen_df_grads.dropna(subset=['Value'])

gen_df_grads.reset_index(drop=True, inplace=True)
gen_df_grads
```
